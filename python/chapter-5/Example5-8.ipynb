# Example 5.8 (Creating a univariate control chart)

```python
import pathlib
import scipy.stats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def load_data() -> pd.DataFrame:
    fpth = pathlib.Path(r'..\..\data\Table5.8.xlsx')
    return pd.read_excel(fpth)
```

```python
df = load_data()
```

```python
x1 = df.iloc[:, 0]
```

```python
xbar1 = np.mean(x1)
s11 = np.var(x1, ddof=1)
```

```python
cl = xbar1 + np.array([-1, 1]) * (3 * np.sqrt(s11))
```

```python
plt.figure()
ax = plt.gca()
plt.axhline(y=cl[0], color='r', linestyle='--')
plt.axhline(y=xbar1, color='g', linestyle='--')
plt.axhline(y=cl[1], color='r', linestyle='--')
plt.plot(x1, marker='x')
plt.ylim([1500, 5500])
plt.title('Legal Appearances Overtime Hours')
plt.xlabel('Observation Number')
plt.ylabel('Individual Value')
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim()) 
ax2.set_yticks([cl[0], xbar1, cl[1]])
ax2.set_yticklabels([f'LCL = {cl[0]:.1f}', f'$\\bar{{x}} = {xbar1:.1f}$', f'UCL = {cl[1]:.1f}'])
plt.show()
```

```python
# Sample data
# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)

# Confidence intervals (just for illustration)
LCL = 1737
UCL = 5379

# Create the figure and the first y-axis
fig, ax1 = plt.subplots()

# Plot data for the first y-axis
ax1.plot(x, y1, 'g-', label='sin(x)')
ax1.set_xlabel('X data')
ax1.set_ylabel('sin(x)', color='g')
ax1.tick_params(axis='y', labelcolor='g')

# Create the second y-axis (twin axis)
ax2 = ax1.twinx()

# Manually set the limits and ticks for the second y-axis
ax2.set_ylim([0, 6000])  # Set the y-limits according to LCL and UCL
ax2.set_yticks([LCL, UCL])  # Set the ticks at LCL and UCL

# Customize the tick labels
ax2.set_yticklabels([f'LCL = {LCL}', f'UCL = {UCL}'])

# Show grid lines for LCL and UCL on the secondary axis
ax2.axhline(LCL, color='blue', linestyle='--', linewidth=1)
ax2.axhline(UCL, color='blue', linestyle='--', linewidth=1)

# Show the plot
plt.show()
```
